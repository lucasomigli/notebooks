```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

x = np.linspace(-10,10)
y = np.linspace(-10,10)
z = np.linspace(-10,10)

def objective(var):
    x=var[0]
    y=var[1]
    z=var[2]
    return x + 2*y - 2*z

def constraint (var):
    x=var[0]
    y=var[1]
    z=var[2]
    return x**2 + 2*y**2 + 4*z**2 -1

f = objective([x, y, z])
g = constraint([x, y, z])
cons = ({'type': 'ineq', 'fun': constraint})

x0 = np.array([0, 0, 0])

plt.figure(figsize=(30,12))

plt.plot(f, linewidth=4)
plt.plot(g, linewidth=4)
plt.show()
```

```python
minimize(objective,
        x0=x0,
        constraints=cons,
        method='SLSQP',
        options={'disp': True})
```
